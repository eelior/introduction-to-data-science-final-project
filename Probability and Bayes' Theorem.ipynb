## Probability and Bayes' Theorem

### Q1.1

- About 1/125 of births are non-identical twins and 1/300 of births are identical twins. Elvis had a twin brother who died in childbirth. What is the probability that Elvis was identical twin? (It can be assumed that the probability of having a son and a daughter is equal to 1/2)

For more comfort writing we will mark:

$ I = Identical $

$ T = Twin/s $

$ N = Not $

$ M = Male $

$ F = Female $

$ B = Brother $

Probabilty for Identical Male Twins (Twin options: MM\FF):

$ P(IMT) = P(M) * P(IT) = \frac{1}{2} * \frac{1}{300} = \frac{1}{600} $

Probabilty for Not Identical Male Twins (Twin options: MM\FF\MF\FM):

$ P(NIMT) = P(M) * P(NIT) = \frac{1}{4} * \frac{1}{125} = \frac{1}{500} $

Probabilty for Male Twins:

$ P(MT) = P(IMT) + P(NIMT) = \frac{1}{600} + \frac{1}{500} $

$ P(IMT|MT) = \frac{P(IMT) * P(MT|IMT)}{P(MT)} = \frac{\frac{1}{600} * 1}{\frac{1}{600} + \frac{1}{500}} = \frac{5}{11} = 0.4545... $

### Q1

- Roi is playing a dice game with Yael. Roi will roll 2 six-sided dice, and if the sum of the dice is divisible by 3, he will win \\$6. If the sum is not divisible by 3, he will lose \\$3. What is Roiâ€™s expected value of playing this game?

There is $ 6^2 $ different combinations to rolling a pair dices.
From them, 12 combinations will give us a sum that divides by 3.
So $ \frac{12}{6^2} = \frac{1}{3} $ is the probability of winning \\$3, and completing it, $ 1 - \frac{1}{3} = \frac{2}{3} $ is the probability of losing \\$3.

So Roi's expected value of playing this game:

$ E(x) = P(win) * 6 + P(lose) * (-3) = \frac{1}{3} * 6 + \frac{2}{3} * (-3) = 0 $
