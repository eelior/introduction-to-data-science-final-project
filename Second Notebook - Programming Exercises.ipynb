### Q1

- Write a program in Python that gets a number in base 10 and prints the number in all the other bases (base 2, base 8 and base 16).

```python
def deciamlToOtherBases(origin): 
    def toLetter(num):
        if (num >= 0 and num <= 9): 
            return chr(num + ord('0')); 
        else: 
            return chr(num - 10 + ord('A')); 
    
    base = [2,8,16];
    res = ["","",""];
    
    for i in range(len(base)):
        num = origin;
        while (num > 0): 
            res[i] = toLetter(num % base[i]) + res[i]; 
            num = int(num / base[i]);
    
    print(f"""
        The representation of {origin} for bases:
        binary - {res[0]}
        Octal - {res[1]}
        hexadeciaml - {res[1]}
    """);
```
